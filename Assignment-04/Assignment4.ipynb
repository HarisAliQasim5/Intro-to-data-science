```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import cityblock
import math
import pandas as panda
import numpy as num
```

```python
# Storing Strings

document1= "data science is one of the most important courses in computer science"
document2= "this is one of the best data science courses"
document3= "the data scientists perform data analysis"
```

```python
# Make Array of All Strings

Corpus = [document1, document2 , document3]
```

```python
Vector = CountVectorizer()
CorpusVectorMatrix = Vector.fit_transform(Corpus)
print(CorpusVectorMatrix)
```

```python
# Tokenization

CorpusTokenization = Vector.get_feature_names_out()
CorpusTokenization
```

```python
CorpusVectorMatrix.toarray()
```

```python
# BOW Table

BowTable = panda.DataFrame(data = CorpusVectorMatrix.toarray(), columns = CorpusTokenization)
BowTable['Total'] = BowTable.sum(axis=1)
BowTable
```

```python
# Term Frequecy

TermFrequency = BowTable.div(BowTable.sum(axis=1), axis=0)
TermFrequency
```

```python
# Inverse Document Frequency

No_of_Doccuments = BowTable.shape[0]
No_of_Word_Occurence= (BowTable > 0).sum(axis=0)
Inverse_Doc_Freq= num.log10(No_of_Doccuments / No_of_Word_Occurence) + 1

Inverse_Doc_Freq
```
