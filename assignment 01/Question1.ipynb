```python
# Sep 17 ,2023
# CSC461 – Assignment1 – Web Scraping
# Muhammad Haris Ali
# Question 01
import requests
import time
from bs4 import BeautifulSoup
import pandas
```

```python
links = ['https://www.imdb.com/title/tt10172266/?ref_=hm_tpks_tt_i_2_pd_tp1_pbr_ic','https://www.imdb.com/title/tt23546646/?ref_=fn_al_tt_5','https://www.imdb.com/title/tt9389998/?ref_=fn_al_tt_2','https://www.imdb.com/title/tt10172266/?ref_=fn_rvi_tt_i_6','https://www.imdb.com/title/tt1517268/?ref_=hm_tpks_tt_i_3_pd_tp1_pbr_ic']
abc = {
 'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.1000.0 Safari/537.36',
 'Accept-Language': 'en-US,en;q=0.9',
 'Accept-Encoding': 'gzip, deflate, br',
 'Connection': 'keep-alive',
 'Cache-Control': 'max-age=0',
 'Upgrade-Insecure-Requests': '1',
}
```

```python
resp = []
for i in links :
  response = requests.get(i , headers=abc)
  if response.status_code == 200:
    a = BeautifulSoup(response.text , 'html5lib')
    name = a.find('span' ,attrs={'class' : 'sc-afe43def-1 fDTGTb'}).text
    raiting = a.find('span' ,attrs={'class' : 'sc-bde20123-1 iZlgcd'}).text
    resp.append({
    'name' : name,
    'stars':raiting
    })
  time.sleep(3)
```

```python
resp
```

```python
excel= pandas.DataFrame(resp)
```

```python
excel_file_path= "movie_scraping.xlsx"
excel.to_excel(excel_file_path, index=False)
```
