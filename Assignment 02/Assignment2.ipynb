```python
with open("example-text.txt",'r') as file :
  textData=file.read()
```

```python
textData
```

```python
# Question 01
re.findall(r'\w+',textData)
```

```python
# Question 02
re.findall(r'[A-Z]\w+',textData)
```

```python
# Question 03
re.findall(r'\w{5}',textData)
```

```python
# Question 04
re.findall(r'"([^"]+)"',textData)
```

```python
# Question 5
re.findall(r'[AEIOUaeiou]\w+',textData)
```

```python
# Question 06
re.findall(r'\w{2}e',textData)
```

```python
# Question 07
re.findall(r'b\w+b',textData)
```

```python
# Question 08
re.sub('[\,\"\",\.\;]',"",textData)
```

```python
# Question 09
re.sub('n\'t'," not",textData)
```

```python
# Question 10
re.sub('\n',"", textData)
```
